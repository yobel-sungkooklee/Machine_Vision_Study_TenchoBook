# 필요한 라이브러리 불러오기

```python
import math #수학 패키지 import
import torch #파이토치 모듈 import
import matplotlib.pyplot as plt #시각화 라이브러리
```

# ❶ -pi부터 pi사이에서 점을 1000개 추출

```python
#-pi부터 pi 사이에서 점을 1000개 추출
x = torch.linspace(-math.pi, math.pi, 1000)
```

# ❷ 실제 사인곡선에서 추출한 값으로 y 만들기

```python
y = torch.sin(x)
```

# ❸ 임의의 가중치(계수)를 뽑아 예측용 y 만들기

```python
a = torch.randn(())
b = torch.randn(())
c = torch.randn(())
d = torch.randn(())
```

# 사인 함수를 근사할 3차 다항식 정의

```python
y_random = a * x**3 + b * x**2 + c * x + d
```

# 그래프 출력하기

```python
# ➍ 실제 y값으로 사인곡선 만들기
plt.subplot(2, 1, 1)
plt.title("y true")
plt.plot(x, y)

# ➎ 임의의 가중치로 만든 예측용 y값으로 사인곡선 만들기

plt.subplot(2, 1, 2)
plt.title("y radnom")
plt.plot(x, y_random)

# ➏ 그래프 출력하기
plt.show()
```

# 학습 전후 비교해보기
